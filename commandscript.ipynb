```python
import pandas as pd
import numpy as np
```

### Menu selection
#### Main menu

```python
form = "n" #main menu type selection
spsfile = "sd" #sps file relative to path
valencestart = 1
valence1 = valencestart 
valence2 = valencestart 
valencenum1max = 12  
valencenum2max = 12  
if ((valence1 + valence2) % 2) == 0:
    xjnz = 0
else:
    xjnz = 1
intfile1 = "usdb" #int file relative to path
#intfile2 = ''  #set off because we arent using second interactions right now but we can simply add in as many as we want actually
scaling = "1, 1, 1, 1" #default scaling
```

#### Case selection

```python
variablevalence = True #do we want to vary our valence values
variablelanczos = True #do we want to vary our lanczos nkeep
#how do we want to divy up the batch
resolution1 = True #determine batch # by incrementing in a fixed difference between the two nkeeps
resolution2 = False #determine batch # by incrementing across the max nkeep
```

#### Lanczos menu

```python
formlanczos = "ld" #lanczos menu type selection
MINlower =  20 #mininum nkeep 1st selection
MAXlower = 200 #maximum nkeep 1st selection
MINupper = 200 #minimum nkeep 2nd selection
MAXupper = 380 #maximum nkeep 2nd selection
store1a = MINlower
store1b = MINupper
difference = MAXupper-MAXlower 
if difference < 160:
    print("values below 160 give no convergence")
increment = 20 #how much we're going to go across   
if resolution1 == True and resolution2 == False:
    batchnummax = difference // increment
elif resolution2 ==true and resolution1 == False:
    batchnummax = MAXupper // increment
else:
    print("improper resolution selection")
print(f"# of iterations = {batchnummax}")
print("../bigstick.x", f"batch{i}_{o}", form, spsfile, valence1, valence2, xjnz, intfile1, "end", scaling, formlanczos, store1a, store1b)
```

### Dataframe creation

```python
keeper = {}
inserter = []
commands = []

#iterate for larger values of lanczos
lala = pd.DataFrame()
for i in "abcdefghijklmnopqrstuvwxyz":
    for o in range(1, batchnummax+1):
        if (valence1 + valence2) % 2 == 0:
            xjnz = 0
        else:
            xjnz = 1
        inserter = ["../bigstick.x", f"batch{i}_{o}", form, spsfile, valence1, valence2, xjnz, intfile1, "end", scaling, formlanczos, store1a, store1b]
        commands.extend(inserter)
        keeper[i] = inserter
        outfile = pd.DataFrame.from_dict(keeper[i])
        bee = pd.concat([lala,outfile], axis=0, ignore_index = True)
        lala = bee
        #print(bee)
        store1a = store1a + increment
        store1b = store1b + increment
        if variablevalence == True:
            if valence1 != valencenum1max:
                valence1 = valence1 + 1
            if valence1 == valencenum1max :
                valence2 = valence2 + 1
        if store1b == MAXupper and store1b == MAXlower:
            break
    store1a = MINlower
    store1b = MINupper
    valence1 = valencestart
    valence2 = valencestart
```

### Output

```python
print(bee)
bee.to_csv("outfile.csv", sep = ' ', header = False, index = False)
```
