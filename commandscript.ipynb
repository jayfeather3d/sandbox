made in python because i know how to use pandas for creating files. below mark range of parameter + setting. 

```python
form = "n" #i, n, m, etc (from main menu)
#output name is handled by script below + path is relative to place of generation
spsfile = "sd"
variablevalence = True
valencestart = 1
valence1 = valencestart #what we will start at
valence2 = valencestart #what we will start (for second valence input)
valencenum1max = 4 #1stvalence maximum 
valencenum2max = 4  #2ndvalence maximum
xjnz = 2 #!!!! <- this will need to be made mathematically proportional to valence1 and valence2
intfile1 = "usdb" 
#intfile2 = ''  #set off because we arent using second interactions right now
scaling = "1, 1, 1, 1" #def value 1, 1, 1, 1
formlanczos = "ld" #ld, ls, ex, etc (from lanczos menu)
#Lanczos algorithm portion
MINlower =  20
MAXlower = 180
MINupper = 180
MAXupper = 380
difference = MAXupper-MAXlower
increment = 20  
#handle outputs
batchnummax = 15 #number of iterations to make in a batch
#(^this needs to be matched to lanczos algorithm portion, as in, it needs to increment proportionally)
```

```python
keeper = {}
inserter = []
commands = []
store1a = MINlower
store1b = MINupper
#iterate for larger values of lanczos
lala = pd.DataFrame()
for i in "abcdefghijklmnopqrstuvwxyz":
    for o in range(1, batchnummax+1):        
        inserter = ["../bigstick.x", f"batch{i}_{o}", form, spsfile, valence1, valence2, xjnz, intfile1, "end", scaling, formlanczos, store1a, store1b]
        commands.extend(inserter)
        keeper[i] = inserter
        outfile = pd.DataFrame.from_dict(keeper[i])
        bee = pd.concat([lala,outfile], axis=0, ignore_index = True)
        lala = bee
        #print(bee)
        store1a = store1a + increment
        store1b = store1b + increment
        if variablevalence == True:
            if valence1 != valencenum1max:
                valence1 = valence1 + 1
            if valence1 == valencenum1max :
                valence2 = valence2 + 1
        if store1b == MAXupper and store1b == MAXlower:
            break
    store1a = MINlower
    store1b = MINupper
    valence1 = valencestart
    valence2 = valencestart
commandtracker = pd.DataFrame(commands)
commandtracker.to_csv("commandtracker.csv", header = False, index = False)
```
